#### 2. Conversão automática de fim de linha para usuários Windows

Se você for um usuário do **Windows**, você **DEVE** configurar o git para que os caracteres de fim de linha sejam convertidos do formato do Windows para o formato usado no Linux e Mac, e vice-versa. Para fazer isso, use o seguinte comando:

```
git config --global core.autocrlf true
```

#### 3. Usuários Linux e Mac

Para garantir a uniformidade de representação do caracter de fim de linha, frequentemente usamos o formato do Linux e Mac. Contudo, pode ser que algum arquivo ainda tenha sido introduzido com o formato do Windows. Assim, devemos permitir que o git corrija automaticamente o formato no repositório. Para isso, você deve configurar o git com o seguinte:

```
git config --global core.autocrlf input
```

````python
# exemplo tirado de https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html
from numpy.random import normal
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
````

````python
# A forma mais simples de juntar diferentes dados e concatenacao
A = pd.Series(["A{}".format(a) for a in range(4)],
              index=range(4), name="A")
B = pd.Series(["B{}".format(a) for a in range(4)],
              index=range(4), name= "B")
C = pd.Series(["C{}".format(a) for a in range(5)],
              index=range(5), name= "C")
pd.concat([A,B,C],axis=1)
````
